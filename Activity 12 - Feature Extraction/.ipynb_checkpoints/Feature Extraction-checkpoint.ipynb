```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from skimage.color import rgb2lab, lab2rgb
from matplotlib import patches as patch
from skimage import measure
from mpl_toolkits.mplot3d import Axes3D
import math
```

# Pre-Processing

```python
def objects(path):
    object  = os.listdir(path)
    return object

name = objects(path = "./Fruits")

def dataset_loader(datatype, imagetype):
    def loadImages(path = "./Fruits/" +str(datatype)+"/"):

        return[os.path.join(path,f) for f in os.listdir(path) if f.endswith(imagetype)]

    filenames = loadImages()
    images = []
    for file in filenames:
        image = cv2.imread(file)
        images.append(image)

    return images
```

```python
for fruit in name:
    plt.figure(dpi = 120, edgecolor = 'k')
    images = dataset_loader(fruit, ".jpg")
    for i in range(len(images)):
        plt.subplot(1, len(images),i+1)
        plt.imshow(images[i][:,:,::-1])
        plt.xticks([])
        plt.yticks([])

    plt.show()
    
```

# Feature Extraction

## Color (CIELab Colorspace)

```python
def LAB_mean(datatype):
    images = dataset_loader(datatype, ".jpg")

    a_ = []
    b_ = []
    hue = []
    for i in range(len(images)):
        img = images[i][:,:,::-1]
        lab = rgb2lab(img)
        a,b = lab[:,:,1].ravel(),lab[:,:,2].ravel()
        a_.append(np.mean(a))
        b_.append(np.mean(b))
        h = math.atan2(np.mean(b),np.mean(a)) * 180 / np.pi
        if h < 0:
            h = 180 + np.abs(h)
        hue.append(h)
        
    return a_, b_, hue
```

## Eccentricity

```python

def Ecc(datatype):
    images = dataset_loader(datatype, ".jpg")
    E = []
    for i in range(len(images)):
        img = images[i]
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.bitwise_not(img)
        thresh = 0.05
        img = img/255
        img[img<thresh] = 0
        img[img>thresh] = 1
        A = measure.label(img, neighbors=4)
        for region in measure.regionprops(A,coordinates='xy'):
            E.append(region.eccentricity)
            
    return E
```

```python
%matplotlib inline
datatype = "Banana"
images = dataset_loader(datatype, ".jpg")
plt.figure(dpi = 120, figsize = [20,2])

for i in range(len(images)):
    plt.subplot(1,len(images),i+1)
    img = images[i][:,:,::-1]
    lab = rgb2lab(img)
    a,b = lab[:,:,1].ravel(),lab[:,:,2].ravel()
    a_ = np.mean(a)
    b_ = np.mean(b)
    plt.plot(a,b,'r,')
    plt.plot(a_,b_,'bo')
    plt.xlim(-80,80)
    plt.ylim(-80,80)
    plt.axhline(0, color = 'k', alpha = 0.5)
    plt.axvline(0, color = 'k', alpha = 0.5)
    plt.title(datatype + str(i+1))
plt.show()

plt.figure(dpi = 200, figsize = [15,2])
E = []
for i in range(len(images)):
    plt.subplot(1,len(images),i+1)

    img = images[i]
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.bitwise_not(img)
    thresh = 0.05
    img = img/255
    img[img<thresh] = 0
    img[img>thresh] = 1
    A = measure.label(img, neighbors=4)
    for region in measure.regionprops(A,coordinates='xy'):
        y, x = region.centroid
        plt.plot(y,x, 'ro')
        plt.imshow(img)
        E.append(region.eccentricity)
        plt.xlabel("E = " + str(np.round(region.eccentricity,3)))
plt.show()
        
```

# Perceptron

```python
def step(a):
    if a >= 0: return 1
    if a < 0: return -1
```

```python

Pair = ["Banana", "Blue Berry"]
d = [1, -1]

fruit = 1
a_, b_, hue = LAB_mean(Pair[fruit])
ecc = Ecc(Pair[fruit])
```

```python
mu = 0.01
np.random.seed(69)
weights = np.random.random_sample(3,)
print(weights)
i = 0
iteration = 0
while iteration < 1000:
    feature = np.array([1,hue[i]/360, ecc[i]])
    a = sum(weights*feature)
    z = step(a)
    delta_w = mu*(d[fruit]-z)*feature
    weights += delta_w
    if d[fruit]-z == 0: 
        break
    else: 
        iteration += 1
#     plt.plot(iteration, weights[0], 'r.')
#     plt.plot(iteration, weights[1], 'g.')
#     plt.plot(iteration, weights[2], 'b.')
#plt.show()
print(weights)
  
```

```python
C = -weights[0]
A = weights[1]
B = weights[2]
m = -A/B
b = C/B
```
