```python
import numpy as np
import matplotlib.pyplot as plt
#import cv2 as cv
from skimage import morphology as mp
```

```python
shapes = []
```

```python
square = np.zeros(shape = [11,11])
square[3:8, 3:8] = 1
plt.imshow(square, extent = [0,10,0,10], origin = "lower")
plt.grid()
```

```python
triangle = np.zeros(shape=[11,11])
triangle[6,4:8] = 1
triangle[5,4:7] = 1
triangle[4,4:5] = 1
plt.imshow(triangle, extent = [0,10,0,10], origin = "lower")
plt.grid()
```

```python
hollow = np.zeros(shape=[15,15])
#hollow = np.z(shape=[10,10])
hollow[2:13, 2:13] = 1
hollow[4:11,4:11] = 0
plt.imshow(hollow, extent = [0,10,0,10], origin = "lower")
```

```python
cross = np.zeros(shape=[11,11])
cross[3:8,5] = 1
cross[5,3:8] = 1
plt.imshow(cross, extent = [0,10,0,10], origin = "lower")
#plt.grid()
```

```python
shapes.append(square)
shapes.append(triangle)
shapes.append(hollow)
shapes.append(cross)
```

```python
p = []
patch1 = np.zeros(shape = [4,4])
patch1[1:3,1:3] = 1
patch2 = np.zeros(shape = [4,4])
patch2[1:3,1] = 1
patch3 = np.zeros(shape = [4,4])
patch3[1,1:3] = 1
patch4 = np.zeros(shape=[5,5])
patch4[2,1:4] = 1
patch4[1:4,2] = 1
patch5 = np.zeros(shape = [4,4])
patch5[1,2],patch5[2,1] = 1,1

p.append(patch1)
p.append(patch2)
p.append(patch3)
p.append(patch4)
p.append(patch5)
```

```python
plt.figure(figsize=(2,8))
for i in range(4):
    plt.subplot(4,1,i+1)
    plt.imshow(shapes[i], cmap = "hot")
    plt.axis("off")
```

```python
plt.figure(figsize=(10,2))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(p[i], cmap = "hot")
    plt.axis("off")
```

```python
erosion = []
for i in range(len(shapes)):
    for j in range(len(p)):
        erosion.append(mp.erosion(shapes[i],p[j]))
        
fig, axes = plt.subplots(4,5, figsize=(10,8), sharex=True, sharey=True, dpi=100)
for i, ax in enumerate(axes.ravel()):
    ax.imshow(erosion[i], extent = [0,5,0,5], origin = "upper")
    ax.axis("off")
    
```

```python
dilation = []
for i in range(len(shapes)):
    for j in range(len(p)):
        dilation.append(mp.dilation(shapes[i],p[j]))
        
fig, axes = plt.subplots(4,5, figsize=(10,8), sharex=True, sharey=True, dpi=100)
for i, ax in enumerate(axes.ravel()):
    ax.imshow(dilation[i], extent = [0,5,0,5], origin = "upper")
    ax.axis("off")
```
