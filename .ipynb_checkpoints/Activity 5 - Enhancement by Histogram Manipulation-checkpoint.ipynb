```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#img = cv2.imread('scene2.png')
#img = cv2.imread('portrait.png')

img = cv2.imread('portrait_lowcontrast.png')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(gray, cmap = 'gray')
```

```python
cdf_line = np.round(np.linspace(0,1,256),2)
cdf_sig = np.round(sigmoid(np.linspace(-7,7,256)),2)

def backprojection(gray_value, cdf_desired):
    T_r = cdf_normalized[gray_value]
    a = cdf_desired.tolist()
    new_gray_value = a.index(np.round(T_r, 2))
    return new_gray_value

def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def Contrast_Stretching(gray):
    Imin = np.amin(gray)
    Imax = np.amax(gray)
    contrast_stretched = (gray-Imin)/(Imax-Imin)
    contrast_stretched = contrast_stretched*255
    contrast_stretched =  contrast_stretched.astype(int)

    gray = gray
    plt.figure(figsize=[15,5])
    plt.subplot(1,2,1)
    plt.imshow(gray, cmap='gray', vmin = 0, vmax = 255)
    plt.subplot(1,2,2)
    plt.imshow(contrast_stretched, cmap='gray', vmin = 0, vmax = 255)
    plt.show()
    
    hist,bins = np.histogram(gray.flatten(),256,[0,256])
    cdf = hist.cumsum()
    cdf_normalized = cdf/cdf.max() #* float(hist.max()) / cdf.max()
    
    hist_,bins_ = np.histogram(contrast_stretched.flatten(),256,[0,256])
    cdf_ = hist_.cumsum()
    cdf_normalized_ = cdf_/cdf_.max()
    
    plt.figure(figsize=[15,4])
    plt.subplot(1,2,1)
    plt.hist(gray.flatten(),256,[0,256], color = 'r')
    plt.hist(contrast_stretched.flatten(),256,[0,256], color = 'g')
    plt.xlim(0,256)
    plt.subplot(1,2,2)
    plt.plot(cdf_normalized, 'r')
    plt.plot(cdf_normalized_, 'g')
    plt.xlim(0,256)
    plt.show()
    
    return contrast_stretched, cdf_normalized
    

def histogram_back_projection(gray, cdf_desired):
    contrast_stretched, cdf_normalized = Contrast_Stretching(gray)
    hbj = np.zeros(shape = np.shape(gray))

    for i in range(np.shape(hbj)[0]):
        for j in range(np.shape(hbj)[1]):
            gray_value = contrast_stretched[i][j]
            hbj[i][j] = backprojection(gray_value, cdf_desired)
            
    hist,bins = np.histogram(gray.flatten(),256,[0,256])
    cdf = hist.cumsum()
    cdf_normalized = cdf/cdf.max()
    
    hist_,bins_ = np.histogram(hbj.flatten(),256,[0,256])
    cdf_ = hist_.cumsum()
    cdf_normalized_ = cdf_/cdf_.max()
    
    plt.figure(figsize=[15,5])
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(gray, cmap='gray', vmin = 0, vmax = 255)
    plt.subplot(1,2,2)
    plt.title('Histogram Manipulated')
    plt.imshow(hbj, cmap='gray', vmin = 0, vmax = 255)
    plt.show()
    
    plt.figure(figsize=[15,5])
    plt.subplot(1,2,1)
    plt.title('Probability Distribution Function')
    plt.hist(gray.flatten(),256,[0,256], color = 'r')
    plt.hist(hbj.flatten(),256,[0,256], color = 'b')
    plt.legend(('original', 'histogram manipulated'))

    plt.subplot(1,2,2)
    plt.title('Cummulative Distribution Function')
    plt.plot(cdf_normalized, 'r')
    plt.plot(cdf_normalized_, 'b')
    #plt.plot(cdf_desired, 'g,')
    plt.legend(('original', 'histogram manipulated'))
    plt.show()
    
```
